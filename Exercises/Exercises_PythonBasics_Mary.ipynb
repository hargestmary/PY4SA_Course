```python
list1 = [80, 90, 95, 87, 82]
def average_fun(numbers):
    return sum(numbers)/len(numbers)
average = average_fun(list1)
print("The average number from the list provided is:", average)
```

## 1.8 Create a script that generates a multiplication table

```python
def multiply(number):
    for i in range (1, 10):
        result = i * number
        print(f"{number} x {i} = {result}") 
multiply(5)
```
